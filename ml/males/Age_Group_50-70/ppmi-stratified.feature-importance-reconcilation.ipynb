```python
classifier_result_dirs = ["LR2", "SVM2", "RF2", "XGBOOST2"]
visits = ["BL", "V02", "V04", "V06", "V08"]
```

```python
import pandas as pd

dataframes = {classifier_result_dir: {visit: None for visit in visits} for classifier_result_dir in classifier_result_dirs}

for classifier_result_dir in classifier_result_dirs:
    for visit in visits:
        feature_importance_file = f"{PATH}/{classifier_result_dir}/feature_importance_data_Male_50-70_{visit}.csv"
        df = pd.read_csv(feature_importance_file)
        dataframes[classifier_result_dir][visit] = df
```

```python
# Find common genes in the 'feature' column across all classifiers for a particular visit
# Create a final dataframe with additional columns for importance values from each classifier

result_dataframes_per_visit = {}

for visit in visits:
    # Find common features across all classifiers for the current visit
    common_features = set.intersection(*[set(dataframes[classifier][visit]['feature']) for classifier in classifier_result_dirs])

    # Create a list to store rows for the resulting dataframe
    result_rows = []
    for feature in common_features:
        row = {'feature': feature}
        # Add importance values for each classifier
        for classifier in classifier_result_dirs:
            df = dataframes[classifier][visit]
            importance_value = df.loc[df['feature'] == feature, 'importance'].values
            row[f"importance_{classifier}"] = importance_value[0] if len(importance_value) > 0 else 0
        result_rows.append(row)

    # Create a dataframe for the visit with the intersected features and importances
    result_dataframes_per_visit[visit] = pd.DataFrame(result_rows)
```
